```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name']=df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#Nettoyage Age des athlètes : Remplacer les données de la colonne CategorieAgesAthletes par leur age (2020 - dateNaissance)
```

```python
df2['Athlete_age']=2020-df2['Athlete year of birth']
```

```python
df2.head(10)
```

```python
#Nettoyage Performance de l'athlète : on va supprimer le "h"
```

```python
df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2['Athlete performance']=df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(10)
```

```python
#On va supprimer les colonnes : Athlete Club , Athlete Country, Athlete year of birth, Athlete Age Category
```

```python
df2=df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
df2.head(10)
```

```python
#Nettoyage des données nulles/manquantes (NaN)
```

```python
df2.isnull().sum()
```

```python
#On a 233 valeurs nulles dans la colonne Athlete Age
```

```python
df2[df2['Athlete_age'].isnull()==1]
```

```python
df2=df2.dropna()
```

```python
df2.isnull().sum()
```

```python
#On a supprimé l'ensemble des coureurs n'ayant pas la donnée Age 
```

```python
#On va rechercher les doublons
```

```python
df2[df2.duplicated() == True]
```

```python
#Réinitialisation de l'index (avec drop = True , ancien index supprimé)
```

```python
df2.reset_index(drop = True)
```

```python
#Corriger les types
```

```python
df2.dtypes
```

```python
#Corriger le type de Athlete_age 
```

```python
df2['Athlete_age'] = df2['Athlete_age'].astype(int)
```

```python
#Corriger le type de Athlete average speed
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
#Corriger le type de Event dates
```

```python
mask = df2['Event dates'].str.contains('-')
```

```python
df2.loc[mask,'Event dates'] = df2.loc[mask,'Event dates'].str.split('-').str.get(1)
```

```python
df2[df2['Event dates'].str.contains('-')]
```

```python
df2['Event dates'] = pd.to_datetime(df2['Event dates'], format='%d.%m.%Y')
```

```python
#Corriger le type de Athlete Performance
```

```python
df2['Athlete performance']=pd.to_timedelta(df2['Athlete performance'])
```

```python
#Rempacer M et F par male et female dans Athlete gender 
```

```python
df2['Athlete gender']=df2['Athlete gender'].replace({'M':'Homme','F':'Femme'})
```

```python
 df2.head(10)
```

```python
#Renommer les colonnes
```

```python
df2.dtypes
```

```python
#Year of event                          int64
#Event dates                   datetime64[ns]
#Event name                            object
#Event distance/length                 object
#Event number of finishers              int64
#Athlete performance          timedelta64[ns]
#Athlete gender                        object
#Athlete average speed                float64
#Athlete ID                             int64
#Athlete_age                            int32
```

```python
df2 = df2.rename(columns = {'Year of event' : 'année_evenement' ,
                             'Event dates' : 'date_evenement',
                             'Event name' : 'nom_evenement',
                             'Event distance/length' : 'distance_evenement',
                             'Event number of finishers' : 'nombre_finisseurs',
                             'Athlete performance': 'performance_athlete', 
                             'Athlete gender' : 'sexe_athlete',
                             'Athlete average speed': 'vitesse_moyenne_athlete',
                             'Athlete ID' : 'id_athlete',
                             'Athlete_age' : 'age_athlete'
                 })
```

```python
df2.head(10)
```

```python
#Ordonner les colonnes
```

```python
df3=df2[['date_evenement', 'nom_evenement', 'distance_evenement', 'nombre_finisseurs', 'id_athlete', 'sexe_athlete', 'age_athlete', 'performance_athlete', 'vitesse_moyenne_athlete']]
```

```python
df3.head(10)
```

```python
#Diagrammes et Graphiques
```

```python
#Evaluer le nombre de coureurs en fonction de la distance (50km/50mi)
```

```python
sns.histplot(df3['distance_evenement'],alpha=0.8,edgecolor='black', color='orange')
```

```python
#Evaluer le nombre de coureurs Homme et Femme en fonction de la distance
```

```python
sns.histplot(df3,x = 'distance_evenement', hue="sexe_athlete", alpha=0.8,multiple="dodge",edgecolor='grey', palette={'Homme' :'green', 'Femme' : 'orange'})
```

```python
sns.displot(df3[df3['distance_evenement'] == '50mi']['vitesse_moyenne_athlete'],alpha=0.8,edgecolor='grey')
```

```python
sns.violinplot(data = df3, x ='distance_evenement',y='vitesse_moyenne_athlete', hue='sexe_athlete',split = True,inner = 'quarts',edgecolor='grey',alpha = 1,palette={'Homme' :'green', 'Femme' : 'orange'})
```
